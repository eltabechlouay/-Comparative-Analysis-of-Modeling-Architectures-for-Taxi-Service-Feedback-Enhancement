```python
'''
1. Low values: Negative values are problematic. Values of zero could be legitimate if the taxi logged a trip that was immediately canceled.

2. High values: The maximum fare amount in this dataset is nearly $1,000, which seems very unlikely. 
High values for this feature can be capped based on intuition and statistics. 
The interquartile range (IQR) is $8. The standard formula of Q3 + (1.5 * IQR) yields $26.50. 
That doesn't seem appropriate for the maximum fare cap. In this case, we'll use a factor of 6, which results in a cap of $62.50.
'''
```

```python
# We need to cancel all of the NEGATIVE numbers:
dt.loc[dt['fare_amount'] < 0, 'fare_amount'] = 0
dt['fare_amount'].min()
```

```python
# Now for the max value:
q1 = dt['fare_amount'].quantile(0.25)
q3 = dt['fare_amount'].quantile(0.75)
iqr = q3 - q1
thresh = q3 + 6 * iqr

dt.loc[dt['fare_amount'] > thresh, 'fare_amount'] = thresh
```

```python
dt['fare_amount'].describe()
```

```python
# Now the duration:
dt['duration'].describe()
```

```python
dt.loc[dt['duration'] < 0, 'duration'] = 0
```

```python
q1 = dt['duration'].quantile(0.25)
q3 = dt['duration'].quantile(0.75)
iqr = q3 - q1
thresh = q3 + 6 * iqr

dt.loc[dt['duration'] > thresh, 'duration'] = thresh
```

```python
# FEATURE ENGINEERING:
```

```python
# Create mean_tripdistance column
''' the model will not know the duration of a trip until after the trip occurs, so you cannot train a model that uses this feature. 
However, you can use the statistics of trips you do know to generalize about ones you do not know.
 - mean_tripdistance will be added where the value at each row is the average for all trips with those pickup and dropoff locations
 - First, we create a helper column called drop_off_location, which contains the unique combination of pickup and dropoff location IDs for each row.
'''
```

```python
# drop_off location:
dt['pick_drop'] = dt['PULocationID'].astype(str) + ' ' + dt['DOLocationID'].astype(str)
```

```python
# First One:
```

```python
group = dt.groupby(['pick_drop']).mean(numeric_only=True)[['trip_distance']]
```

```python
groupdict = group.to_dict()
groupdict = groupdict['trip_distance']
```

```python
'''Next Step:
Now, we use map() method on the mean_distance series. Pass groupdict as its argument. 
Reassign the result back to the mean_distance series. When you pass a dictionary to the Series.map() method, it will 
replace the data in the series where that data matches the dictionary's keys. The values that get imputed are the values of the dictionary.
'''
```

```python
# Mean_distance:
dt['mean_distance'] = dt['pick_drop']
dt['mean_distance'] = dt['mean_distance'].map(groupdict)

# Confirmation
# dt[(dt['PULocationID']==100) & (dt['DOLocationID']==231)][['mean_distance']]
```

```python
# Now for mean_duration:
group1 = dt.groupby(['pick_drop']).mean(numeric_only=True)[['duration']]
```

```python
groupdict1 = group1.to_dict()
groupdict1 = groupdict1['duration']
```

```python
dt['mean_duration'] = dt['pick_drop']
dt['mean_duration'] = dt['mean_duration'].map(groupdict1)

# Confirmation
# dt[(dt['PULocationID']==100) & (dt['DOLocationID']==231)][['mean_duration']]
```

```python
# Again, we need the day and month columns in dt data for pick_up datetime so to work with the rush_hour:
dt['day'] = dt['tpep_pickup_datetime'].dt.day_name().str.lower()
dt['month'] = dt['tpep_pickup_datetime'].dt.month_name().str.lower()
```

```python
# We need to create a rush hour column:
'''2 conditions:
1. Any weekday (not Saturday or Sunday) AND
2. Either from 06:00–10:00 or from 16:00–20:00
'''
```

```python
dt['rush_hour'] = dt['tpep_pickup_datetime'].dt.hour # hour
dt.loc[dt['day'].isin(['saturday','sunday']), 'rush_hour'] = 0
```

```python
def rush(hour):
    if 6 <= hour['rush_hour'] < 10:
        val = 1
    elif 16 <= hour['rush_hour'] < 20:
        val = 1
    else:
        val = 0
    return val
```

```python
dt.loc[(dt.day != 'saturday') & (dt.day != 'sunday'), 'rush_hour'] = dt.apply(rush,axis=1)
```

```python
# Time to train and test the data:
```

```python
x = dt2.drop(columns=['fare_amount'],axis=1)
y = dt2[['fare_amount']]
```

```python
# VendorID is a categorical value as int, so we need it as str:
x['VendorID'] = x['VendorID'].astype(str)
x = pd.get_dummies(x, drop_first=True)
```

```python
import sklearn
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# STandardizing the data:
from sklearn.preprocessing import StandardScaler

scale = StandardScaler().fit(xtrain)
xtrainscaled = scale.transform(xtrain)
```

```python
# Fitting the model:
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(xtrainscaled,ytrain)
```

```python
# Let us evaluate the models for TRAIN DATA:
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

score = lr.score(xtrainscaled,ytrain)
yhat = lr.predict(xtrainscaled)

print('Coefficient of Determination:', score)
print('R^2 of Train:', r2_score(ytrain,yhat))
print('MSE of Train:', mean_squared_error(ytrain,yhat))
print('MAE of Train:', mean_absolute_error(ytrain,yhat))
print('RMSE of Train:', np.sqrt(mean_squared_error(ytrain,yhat)))
```

```python
# Let us evaluate the models for TEST DATA:
xtestscaled = scale.transform(xtest)
score = lr.score(xtestscaled,ytest)
yhat1 = lr.predict(xtestscaled)

print('Coefficient of Determination:', score)
print('R^2 of Train:', r2_score(ytest,yhat1))
print('MSE of Train:', mean_squared_error(ytest,yhat1))
print('MAE of Train:', mean_absolute_error(ytest,yhat1))
print('RMSE of Train:', np.sqrt(mean_squared_error(ytest,yhat1)))
```

```python
'''
The model performance is high on both training and test sets, suggesting that there is little bias in the model and that the model is not overfit. 
In fact, the test scores were even better than the training scores.

For the test data, an R2 of 0.868 means that 86.8% of the variance in the fare_amount variable is described by the model.

The mean absolute error is informative here because, for the purposes of the model, an error of two is not more than twice as bad as an error of one.
'''
```

```python
# We will need the day columns. First, let's convert the data to date of pickup and dropoff:
df1['tpep_pickup_datetime'] = pd.to_datetime(df1['tpep_pickup_datetime'])
df1['tpep_dropoff_datetime'] = pd.to_datetime(df1['tpep_dropoff_datetime'])
```

```python
# Now creating day column:
df1['day'] = df1['tpep_pickup_datetime'].dt.day_name().str.lower()
```

```python
# Now, we create four new columns that represent time of day bins:
'''
am_rush = [06:00–10:00)
daytime = [10:00–16:00)
pm_rush = [16:00–20:00)
nighttime = [20:00–06:00)
'''
```

```python
df1['am_rush'] = df1['tpep_pickup_datetime'].dt.hour
df1['daytime'] = df1['tpep_pickup_datetime'].dt.hour
df1['pm_rush'] = df1['tpep_pickup_datetime'].dt.hour
df1['nighttime'] = df1['tpep_pickup_datetime'].dt.hour
```

```python
# Now,we need the 4 functions as binary values(0,1):
def am_rush(hour):
    if 6 <= hour['am_rush'] < 10:
        val = 1
    else:
        val = 0
    return val

def daytime(hour):
    if 10 <= hour['daytime'] < 16:
        val = 1
    else:
        val= 0
    return val

def pm_rush(hour):
    if 16 <= hour['pm_rush'] < 20:
        val = 1
    else:
        val = 0
    return val

def nighttime(hour):
    if 20 <= hour['nighttime'] < 24:
        val = 1
    elif 0 <= hour['nighttime'] < 6:
        val = 1
    else:
        val = 0
    return val
```

```python
# Now Applying them:
df1['am_rush'] = df1.apply(am_rush,axis=1)
df1['daytime'] = df1.apply(daytime,axis=1)
df1['pm_rush'] = df1.apply(pm_rush,axis=1)
df1['nighttime'] = df1.apply(nighttime,axis=1)
```

```python
# Now for usimg the modelling:
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import GridSearchCV

rf = RandomForestClassifier(random_state=42)

params = {'max_depth': [None], 'max_features':[1.0], 'max_samples':[0.7], 'min_samples_leaf':[1], 'min_samples_split':[2],
          'n_estimators':[300]}

score = {'accuracy':'accuracy','precision':'precision','recall':'recall','f1':'f1'}

rf1 = GridSearchCV(rf,params,scoring=score,cv=4,refit='f1')
```

```python
# We will use the make_results() function to output all of the scores of your model:
def make_results(model_name:str, model_object, metric:str):
    metric_dict = {'precision': 'mean_test_precision',
                   'f1':'mean_test_f1',
                   'recall':'mean_test_recall',
                   'accuracy':'mean_test_accuracy'}

    cv_results = pd.DataFrame(model_object.cv_results_)

    best_estimator_results = cv_results.iloc[cv_results[metric_dict[metric]].idxmax(), :]

    f1 = best_estimator_results.mean_test_f1
    recall = best_estimator_results.mean_test_recall
    accuracy = best_estimator_results.mean_test_accuracy
    precision = best_estimator_results.mean_test_precision

    table = pd.DataFrame({'model':[model_name],
                          'precision':[precision],
                          'recall':[recall],
                          'f1':[f1],
                          'accuracy':[accuracy]})
    return table
```

```python
# We will get the scores on the test data:
rf_preds = rf1.best_estimator_.predict(xtest)
```

```python
# Now make results ofr test:
def make_results_test(model_name:str, predictions, testdata_of_y):
    
    accuracy = accuracy_score(testdata_of_y, predictions)
    recall = recall_score(testdata_of_y, predictions)
    f1 = f1_score(testdata_of_y, predictions)
    precision = precision_score(testdata_of_y, predictions)

    table = pd.DataFrame({'model':[model_name], 
                          'precision':[precision], 
                          'recall':[recall], 
                          'f1':[f1], 
                          'accuracy':[accuracy]})
    return table
```
